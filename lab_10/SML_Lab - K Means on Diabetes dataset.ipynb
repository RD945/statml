#K-means clustering on Diabetes Dataset

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the diabetes dataset
data = pd.read_csv('diabetes.csv')
```

```python
if 'Outcome' in data.columns:
    data = data.drop(columns=['Outcome'])
```

```python
# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
# KMeans clustering
# Define the number of clusters (e.g., 2 or 3)
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(scaled_data)
```

```python
# Get the cluster labels
data['Cluster'] = kmeans.labels_
```

```python
# Optional: Use PCA for 2D visualization
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
# Plotting the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=data['Cluster'], palette="viridis")
plt.title('K-means Clustering on Diabetes Dataset')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.show()
```
