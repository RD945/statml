1. Import matplotlib and pandas libraries.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

2. Load the given dataset - ‘ecommerce_data.csv’.


```python
data = pd.read_csv("ecommerce_data.csv")
print(data.head())
```

3. Create a simple line plot in Matplotlib to show total profit of the ecommerce website month wise.
Use the Month column of the data on the x-axis and total profit on the y-axis

```python
months = data['month_number']
total_profit = data['total_profit']
```

```python
plt.figure(figsize=(5, 3))
plt.plot(months, total_profit, marker='o', linestyle='-', color='b')
plt.grid()
plt.show()
```

4. Using the same plot as in Q3. Add the following style properties in the graph:


a. Line should be red-colored.

```python
plt.figure(figsize=(5, 3))
plt.plot(months, total_profit, marker='o', linestyle='-', color='r')
plt.grid()
plt.show()
```

b. Each axis should be properly labeled.


c. Display title of the plot.


```python
plt.figure(figsize=(5, 3))
plt.plot(months, total_profit, marker='o', linestyle='-', color='r')
plt.title('Month-wise Total Profit of the E-commerce')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid()
plt.show()
```

d. Show legend at the lower right location.


```python
plt.figure(figsize=(5, 3))
plt.plot(months, total_profit, marker='o', linestyle='-', color='r', )
plt.title('Month-wise Total Profit of the E-commerce')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.legend(['Total Profit'], loc='lower right')
plt.grid()
plt.show()
```

e. Line width should be 3.

```python
plt.figure(figsize=(5, 3))
plt.plot(months, total_profit, marker='o', linestyle='-', color='r', linewidth=3 )
plt.title('Month-wise Total Profit of the E-commerce')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.legend(['Total Profit'], loc='lower right')
plt.grid()
plt.show()
```

5. For each month, plot the 'shirts', 't shirts', 'trousers', and 'shoes' data in one graph. Add the
following style properties in the graph:

a. Each line should be of a different color.


b. Set the linewidth to be 3.


c. Set marker as 'o'.


d. Set the titles for x and y label axes.


e. Set the title of the graph.


f. Set the legend in the upper right corner inside the plot.


```python
shirts = data['shirts']
t_shirts = data['t shirts']
trousers = data['trousers']
shoes = data['shoes']
```

```python
plt.figure(figsize=(7, 3))
plt.plot(months, shirts, marker='o', linestyle='-', color='b', linewidth=3, label='Shirts')
plt.plot(months, t_shirts, marker='o', linestyle='-', color='g', linewidth=3, label='T-Shirts')
plt.plot(months, trousers, marker='o', linestyle='-', color='r', linewidth=3, label='Trousers')
plt.plot(months, shoes, marker='o', linestyle='-', color='gold', linewidth=3, label='Shoes')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data for Different Product Categories')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

Section 2: Seaborn

Q1. Import seaborn library.

```python
import seaborn as sns
```

Q2. Load the built-in 'tips' dataset in seaborn and check the columns.


```python
tips = sns.load_dataset('tips')
print(tips.columns)
```

Q3. Plot the data distribution of the column 'total_bill'.


```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'],bins=30)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill')
plt.show()
```

Q4. Use the bar plot for 'total bill' and 'day' columns.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='total_bill', data=tips, estimator='mean')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill')
plt.title('Average Total Bill per Day')
plt.show()
```

Q5. Draw Heatmap by finding the correlation of numerical columns in the data.


```python
num_cols= tips.select_dtypes(include=['number'])
corr = num_cols.corr()
plt.figure(figsize=(6,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidth= 0.5,linecolor= 'black')
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```
