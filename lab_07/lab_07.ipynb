```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

SECTION-1
1: Load the IRIS dataset and display the first few rows.

```python
iris=load_iris()
iris.data.shape
```

```python
iris.data.tolist()
```

```python
iris.target
```

```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris.head()
```

Q2: Split the dataset into training and testing sets.

```python
x = df_iris.drop('species', axis=1)
y = df_iris['species']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Q3: Standardize the features for KNN using Standard Scalar.

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Q4: Train a K-Nearest Neighbor model with K=3 on the training data.

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

Q5: Display the classification report.

```python
from sklearn.metrics import classification_report

y_pred = knn.predict(x_test)
print(classification_report(y_test, y_pred))
```

Q6: Train a Gaussian Naive Bayes model on the training data.

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

Q7: Display the classification report.

```python
y_pred = gnb.predict(x_test)
print(classification_report(y_test, y_pred))
```

Section 2: K-Nearest Neighbor Model and Naïve Bayes on
Diabetes Dataset

Q1: Load the dataset and display the first few rows.

```python
df = pd.read_csv('diabetes.csv')
print(df.head())
```

Q2: Split the dataset into training and testing sets.

```python
X = df.drop('Outcome', axis=1)
Y = df['Outcome']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Q3: Standardize the features for KNN using Standard Scalar.

```python

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Q4: Train a K-Nearest Neighbor model with K=3 on the training data.

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
```

Q5: Display the classification report.

```python
Y_pred = knn.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

Q6: Train a Gaussian Naive Bayes model on the training data.

```python
gnb = GaussianNB()
gnb.fit(X_train, Y_train)
```

Q7: Display the classification report.

```python
Y_pred = gnb.predict(X_test)
print(classification_report(Y_test, Y_pred))
```
